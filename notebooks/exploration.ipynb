```python
import pandas as pd
import os

def explore_csv_files(csv_folder):
    # Get the list of CSV files in the folder
    csv_files = [f for f in os.listdir(csv_folder) if f.endswith('.csv')]

    # Iterate over each CSV file
    for csv_file in csv_files:
        table_name = os.path.splitext(csv_file)[0]  # Extract table name from file name
        csv_file_path = os.path.join(csv_folder, csv_file)

        # Read CSV file into DataFrame
        df = pd.read_csv(csv_file_path)

        # Print table name
        print(f"Table: {table_name}")

        # Print DataFrame information
        print(df.info())

        print()

# Replace 'data' with the path to the folder containing your CSV files
explore_csv_files('data/csv_files')
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read the vintages table into a DataFrame
vintages_df = pd.read_csv('data/csv_files/vintages.csv')

# Select the price_euros column
prices = vintages_df['price_euros']

# Plot the distribution of prices using seaborn
sns.histplot(prices, kde=True)
plt.title('Distribution of Prices from Vintages Table')
plt.xlabel('Price (Euros)')
plt.ylabel('Frequency')
plt.show()
```

```python
import numpy as np

# Apply log transformation to prices
log_prices = np.log(prices)

# Plot the distribution of log-transformed prices using seaborn
sns.histplot(log_prices, kde=True)
plt.title('Distribution of Log-Transformed Prices from Vintages Table')
plt.xlabel('Log(Price)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Create three bins for log-transformed prices
price_bins = pd.cut(log_prices, bins=3, labels=['low', 'medium', 'high'])

# Plot the distribution of log-transformed prices using seaborn
sns.histplot(price_bins, kde=False)
plt.title('Distribution of Log-Transformed Prices (Binned) from Vintages Table')
plt.xlabel('Price Category')
plt.ylabel('Frequency')
plt.show()
```

```python
# Read the wines table into a DataFrame
wines_df = pd.read_csv('data/csv_files/wines.csv')

# Select the ratings_average column
ratings = wines_df['ratings_average']

# Plot the distribution of ratings_average using seaborn
sns.histplot(ratings, kde=True)
plt.title('Distribution of Ratings Average from Wines Table')
plt.xlabel('Ratings Average')
plt.ylabel('Frequency')
plt.show()
```

```python
# Create three bins for log-transformed prices
rating_bins = pd.cut(ratings, bins=3, labels=['low', 'medium', 'high'])

# Plot the distribution of log-transformed prices using seaborn
sns.histplot(rating_bins, kde=False)
plt.title('Distribution of Ratings (Binned) from Wines Table')
plt.xlabel('Average Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
import pandas as pd
import os

# Directory containing CSV files
csv_dir = 'data/csv_files'

# List all CSV files in the directory
csv_files = [file for file in os.listdir(csv_dir) if file.endswith('.csv')]

# Iterate over each CSV file
for file in csv_files:
    # Read the CSV file into a DataFrame
    df = pd.read_csv(os.path.join(csv_dir, file))

    # Identify float-type columns and replace periods with commas
    for column in df.columns:
        if df[column].dtype == 'float64' or df[column].dtype == 'float32':
            df[column] = df[column].astype(str).str.replace('.', ',')

    # Save the modified DataFrame back to the CSV file
    df.to_csv(os.path.join(csv_dir, file), index=False)

    print(f"{file} modified successfully.")
```
